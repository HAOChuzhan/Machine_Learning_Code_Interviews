#### 算法流程
算法流程为（距离定义使用欧几里得距离）：
1. 令S为质心集合，初始化为空集
2. 从数据集中随机选取一个样本点，加入S成为第一个质心
3. for i = 2 to k：  
a. 对于不在S中的每个样本，计算其与当前S中所有质心的距离的最大值D  
b. 选取D值最大的点，加入S成为第i个质心  
（此时S中包含k个质心）
3. 对每个样本点：  
a. 计算其与S中各个质心的距离  
b. 将样本点分配至最近质心所属的类
4. 对某个类：  
a. 将质心设置为该类所有样本点的均值  
b. 在S中更新对应的类的质心
如果S中有质心发生改变，则重复上述过程（4-5）

```python
# 随机产生相关的测试数据集（散点） 
import numpy as np
import random
import itertools
from matplotlib import pyplot as plt
random.seed(1)

n = 50 # 随机产生50个点
random_n_list =list(itertools.product(range(-10,10),range(-10,10)))
random_n_list =  np.array(random.sample(random_n_list, n))
print(random_n_list.shape) # 数据集大小
x = np.array([xy[0] for xy in random_n_list])
y = np.array([xy[1] for xy in random_n_list])

is_used = np.ones([n ,1])

# 绘制随机的采样点
plt.scatter(x, y)
plt.show()

# 定义两点的欧式距离
def distance(pointA, pointB):
    return np.sqrt((pointA[0]-pointB[0])**2+(pointA[1]-pointB[1])**2)
```

```python
def rand_center1(data, k): # 随机初始化聚类中心点
    num_features = data.shape[1] 
    init_centers = np.zeros((k, num_features)) 
    for i in range(num_features): 
        min_data = min(data[:, i])
        max_data = max(data[:, i])
        range_data = max_data - min_data 
        init_centers[:, i] = min_data + range_data * np.random.rand(k) 
    return init_centers

def rand_center2(data, k): # 根据算法流程选择存在的点作为质心
    S = [] # 保存质心的列表
    point_idx = np.random.randint(n)
    rand_point = random_n_list[point_idx]
    is_used[point_idx] = 0 # 被归纳进质心的list中
    S.append(rand_point)
    print(S)
    dis = 0
    max_point = []
    while (len(S)<=k-1):
        for point in random_n_list:
            tmp = 0
#             if point in S:
#                 continue
            for center in S:    
                tmp += distance(center, point)    
            if tmp>dis:
                dis = tmp
                max_point = point
        S.append(max_point)
    return np.array(S)

plt.figure(figsize=(12, 6)) # 画面大小其实是 1200 * 600的。

# 初始的聚类中心点
plt.subplot(131)
plt.title("The initial cluster centers")
plt.scatter(rand_center1(random_n_list, 5)[:, 0], rand_center1(random_n_list, 5)[:, 1])
plt.xlim((-10, 10))
plt.ylim((-10, 10))

def clustering(data, k):
    N, m = data.shape
    cluster_centers = rand_center2(data, k) #是该方法将初始点定位数据内部
    cluster_assign = np.zeros((N, ))#创建数组，用于存放所属类别
    change = True
    while change:
        change = False
        for i in range(N): #遍历每行数据，来计算其与k个聚类中心点的距离，并判定其所属类别
            min_dis = np.inf #初始化一个极大值作为最小距离
            index = 0 #初始化所属类别
            for j in range(k):
                dis = distance(data[i], cluster_centers[j]) #计算当前行与每个距离中心点的距离
                if dis < min_dis:
                    min_dis = dis #重新给最小距离赋值
                    index = j #确定其所属类别
            if cluster_assign[i] != index: 
                change = True
            cluster_assign[i] = index
        
        for j in range(k): #更新聚类中心点
            cluster_data = data[np.nonzero(cluster_assign == j)] 
            if len(cluster_data): 
                cluster_centers[j] = np.mean(cluster_data, axis = 0)
    return cluster_centers, cluster_assign

center_k, res = clustering(random_n_list, 5)

# 展示最后的聚类中心坐标
plt.subplot(132)
plt.scatter(center_k[:, 0], center_k[:, 1])
plt.title("The final cluster centers")
plt.xlim((-10, 10))
plt.ylim((-10, 10))

# 最后所有点的分类结果（颜色区分种类）
plt.subplot(133)
plt.scatter(x, y, c=res)

plt.show()
```
